```python
from pydantic import BaseModel
from typing import Optional
```

```python
class Model(BaseModel):
    id: int 
    first_name: str | None = None
    last_name: str 
```

```python
student1 = Model(id=1, last_name="Deo")
```

```python
student1
```

```python
student1.model_dump(exclude_unset=True)
```

```python
student1.model_fields
```

```python
student1.model_fields_set
```
