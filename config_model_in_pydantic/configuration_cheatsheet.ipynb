# configuration Model 

```python
from pydantic import BaseModel, ConfigDict
```

##### 1) str_to_lower

```python
class Model(BaseModel):
    model_config = ConfigDict(str_to_lower=True)
    first_name : str
    
```

```python
m = Model(first_name="REZA")
```

```python
m
```

```python
m.model_dump()
```

##### 2) str to upper

```python
class Model(BaseModel):
    model_config = ConfigDict(str_to_upper=True)
    first_name: str
```

```python
m = Model(first_name="reza")
```

```python
m.model_dump()
```

##### 3) str strip whitespace

```python
class Model(BaseModel):
    model_config = ConfigDict(str_strip_whitespace=True)
    first_name: str
```

```python
m = Model(first_name="    R    e  Z  A     ")
```

```python
m
```

##### 4) extra "ignore" = default , "forbid" = error , "allow" = acces

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="ignore")
    first_name : str
```

```python
Model.model_validate({"first_name":"freza", "age":20})
```

```python
class Model(BaseModel):
    model_config  = ConfigDict(extra="forbid")
    first_name: str
    
```

```python
Model.model_validate({"first_name":"freza", "age":20})
```

```python
class Model(BaseModel):
    model_config = ConfigDict(extra="allow")
    first_name: str
```

```python
m = Model.model_validate({"first_name":"freza", "age":20})
```

```python
m.model_dump()
```

```python
m.model_fields
```

```python
m.model_fields_set
```

```python
m.model_dump(exclude=m.model_fields)
```

```python
m.model_dump(exclude_unset=True)
```

```python
class Model(BaseModel):
    model_config = ConfigDict(frozen=True)
    first_name: str
```

```python
m = Model(first_name="Reza")
```

```python
m
```

```python
m.first_name="Farshid"
```

```python
m2 = m.model_copy(update={"first_name": "Farshid"})
```

```python
m2
```

```python
m
```

#### 6) populate_by_name

```python
from pydantic import Field
class Model(BaseModel):
    model_config = ConfigDict(populate_by_name=True)
    id: int = Field(alias="iD")


m1 = Model(id=1)
m2 = Model(iD=2)
```
