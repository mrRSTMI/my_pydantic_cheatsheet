```python
from pydantic import BaseModel, ConfigDict , Field
```

```python
class Model(BaseModel):
    model_config = ConfigDict(
        validate_assignment=True,
        validate_by_name=True,
        extra="allow",
        # extra="forbid",
        # extra="ignore"
        str_strip_whitespace=True,
        str_to_lower=True,
        str_to_upper=True
    )
```
