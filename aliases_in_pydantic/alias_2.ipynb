```python
from pydantic import BaseModel, Field, ConfigDict, AliasChoices
```

```python
class Model(BaseModel):
    model_config = ConfigDict(
        populate_by_name=True,
        validate_assignment=True,
        validate_default=True,
    )
    id_: int = Field(
        alias="id",
        validation_alias=AliasChoices("ID", "iD", "Id"),
        serialization_alias="id"
    )
```

```python
m = Model(ID=1)
m1 = Model(iD = 2)
m2 = Model(id_=3)
```

```python

print(m)
print(m1)
print(m2)

```

```python
m.model_dump(by_alias=True)
```

```python
m.model_dump(exclude_unset=True, by_alias=True)
```

```python
from pydantic.alias_generators import to_camel
```

```python
class Model(BaseModel):
    model_config = ConfigDict(
        populate_by_name=True,
        validate_assignment=True,
        validate_default=True,
        alias_generator=to_camel,
    )
    id_: int = Field(
        alias="id",
        validation_alias=AliasChoices("ID", "iD", "Id"),
        serialization_alias="id",
    )
    last_name: str = Field(
        alias="firstName",
        validation_alias=AliasChoices("lName", "LastName", "lastName"),
        serialization_alias="lastName"
    )
```

```python
person1 = Model(ID = 1 , LastName="rostami")
```

```python
person1
```

```python
person1.model_dump(by_alias=True)
```

```python
person2 = Model(ID = 2, lName = "nemati")
```

```python
person2
```

```python
person2.model_dump(by_alias=True)
```
